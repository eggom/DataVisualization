# Section 1: Matplotlib 
In this section, I will go through basic plots with Python.

```python
# Import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')

# importing data analysis library and creating dataset
import numpy as np
```

```python
# Let's create an example dataset
x=np.linspace(0,1,10)
y=x**2
```

```python
x,y
```

```python
# Let's plot y vs x
plt.plot(x,y)
```

# Plot with title, axis name and color

```python
plt.plot(x,y,'r') # 'r' is the color
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple X-Y plot')
plt.show()
```

# Creating multiple plots in the same canvas

```python
# plt.subplot(nrows, ncols, plot_numbers)
plt.subplot(1,2,1) # right most 1 is the plot number
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'b*-')
```

# Object orientated method in creating plots

```python
# create a figure object(empty canvas)
fig=plt.figure()

# add axes to figure
axes=fig.add_axes([0.0,0.0,1.0,1.0])

axes.plot(x,y,'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title of the graph')
```

```python
# Creating subplots
fig=plt.figure()
axes1=fig.add_axes([0.1,0.2,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.3,0.4])

axes1.plot(x,y,'b')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Outer plot')


axes2.plot(x,y,'r*-')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('inset plot')
```

# Specifiying number of columns and rows

```python
fig, axes=plt.subplots( nrows=1, ncols=2)
```

```python
for ax in axes:
    ax.plot(x,y,'r*-')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
    
fig
```

# Get rid of overlapping axes.

```python
fig, axes=plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y,'b*-')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
    
fig

fig.tight_layout()
```

# Setting colors, line widths, line types

```python
fig, ax=plt.subplots(figsize=(4,6), dpi=100)

ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3+1, 'r*-')
```

# Color Parameter

```python
fig, ax=plt.subplots()
ax.plot(x, x**2, color='red', alpha=0.5)# transfarant
ax.plot(x, x+2, color='#8B008B') # RGB hex code
ax.plot(x, x+3, color='#FF8C00')
```

# Line and Marker Style

```python
fig, ax= plt.subplots(figsize=(8,6))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Graph with different style and marker')

ax.plot(x, x+1, color='red', lw=1.5)
ax.plot(x, x+2, color='red', lw=0.5)
ax.plot(x, x+3, color='red', lw=0.9)

#line style
ax.plot(x, x+4, color='blue', lw=1.5, ls='--')
ax.plot(x, x+5, color='blue', lw=1.5, ls='-.')
ax.plot(x, x+6, color='blue', lw=1.5, ls='-')


# marker symble
ax.plot(x, x+7, color='green', lw=1.5, ls='--', marker='o')
ax.plot(x, x+8, color='green', lw=1.5, ls='-.', marker='+')
ax.plot(x, x+9, color='green', lw=1.5, ls='-', marker='1')


# marker marker size
ax.plot(x, x+10, color='y', lw=1.5, ls='-', marker='o', markersize=5)
ax.plot(x, x+11, color='y', lw=1.5, ls='-', marker='o', markersize=8, markerfacecolor='red')
ax.plot(x, x+12, color='y', lw=1.5, ls='-', marker='o', markersize=10)
```

```python
fig, axes=plt.subplots(1,2, figsize=(10,4))

axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('Normal Scale')


axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale('Log')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title('Logarithmic Scale "Y" ')
```

```python
fig, ax= plt.subplots(figsize=(10,4))
ax.plot(x*5,(x*5)**2, x*5, (x*5)**3, lw=2)
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18 )

yticks=[0,50,100,150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%1f$" % y for y in yticks], fontsize=18)
```

# Scientific Notation

```python
fig, ax=plt.subplots(1,1)

ax.plot(x*5, (x*5)**2, x*5, np.exp(x*5))
ax.set_title('Scientific Notation')
ax.set_yticks([0,50,100,150])

from matplotlib import ticker

formatter=ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter) 
```

# Axis Grid

```python
fig, ax=plt.subplots(1,2, figsize=(10,3))

#default grid
ax[0].plot(x*5, (x*5)**2, x*5, np.exp(x*5))
ax[0].grid(True)

ax[1].plot(x*5, (x*5)**2, x*5, np.exp(x*5))
ax[1].grid(color='b', lw=2, ls='dashed', alpha=0.5 )
```

# Twin Axis

```python
fig, ax1=plt.subplots(figsize=(8,6))

ax1.plot(x*5+2, (x*5)**2, lw=2, color='b')
ax1.set_ylabel(r'area $(m^2)$', fontsize=18, color='b')

ax2=ax1.twinx()
ax2.plot(x*5, (x*5)**3, lw=2, color='g')
ax2.set_ylabel(r'volume $(m^3)$', fontsize=18, color='g')
```

# Axes where x and y is zero

```python
fig, ax=plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3);
```
