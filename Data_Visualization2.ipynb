# Plot with Seaborn

# Catergorical Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(20)
```

```python
# bar plot
sns.barplot(x='sex', y='total_bill', data=tips)
```

# Changing estimator

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips )
```

```python
sns.boxplot(x='day', y='tip', data=tips)
```

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill',hue='smoker', data=tips, palette= 'coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker', palette='rainbow')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex', data=tips)
```

# Histogram Plot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=10)
```

# Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

# Pair Plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

# Grid Plot

```python
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.PairGrid(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Pair PLot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species',palette='rainbow')
```

```python
g=sns.FacetGrid(tips,col='time',row='smoker')
```

```python
g=sns.JointGrid(x='total_bill', y='tip', data=tips)
g=g.plot(sns.regplot,sns.distplot)
```

# Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,hue='sex',palette='coolwarm', aspect=0.6, size=8)
```

# Matrix Plot

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# heatmap

tips.corr()
```

```python
flights.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
```

```python
flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
pvtable=flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
sns.heatmap(pvtable)
```

```python
sns.heatmap(pvtable, annot=True)
```

```python
sns.heatmap(pvtable, linecolor='white', linewidths=0.1)
```

# Cluster map

```python
sns.clustermap(pvtable, linecolor='black', linewidths=0.1)
```

```python
sns.clustermap(pvtable, linecolor='black', linewidths=0.1, standard_scale=1)
```
